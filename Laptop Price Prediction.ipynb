```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
dataset['Company'] = encoder.fit_transform(dataset['Company'].values.reshape(-1, 1))
dataset['Product'] = encoder.fit_transform(dataset['Product'].values.reshape(-1, 1))
dataset['TypeName'] = encoder.fit_transform(dataset['TypeName'].values.reshape(-1, 1))
dataset['ScreenResolution'] = encoder.fit_transform(dataset['ScreenResolution'].values.reshape(-1, 1))
dataset['CPU_Company'] = encoder.fit_transform(dataset['CPU_Company'].values.reshape(-1, 1))
dataset['CPU_Type'] = encoder.fit_transform(dataset['CPU_Type'].values.reshape(-1, 1))
dataset['Memory'] = encoder.fit_transform(dataset['Memory'].values.reshape(-1, 1))
dataset['GPU_Company'] = encoder.fit_transform(dataset['GPU_Company'].values.reshape(-1, 1))
dataset['GPU_Type'] = encoder.fit_transform(dataset['GPU_Type'].values.reshape(-1, 1))
dataset['OpSys'] = encoder.fit_transform(dataset['OpSys'].values.reshape(-1, 1))
```

```python
dataset.head()
```

```python
y = dataset["Price (Euro)"].values
x = dataset.drop("Price (Euro)", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score,  mean_squared_error
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
sbn.regplot(x= y_pred, y= y_test, data=dataset)
plt.title('Regression Plot: Predicted vs. Actual')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```
